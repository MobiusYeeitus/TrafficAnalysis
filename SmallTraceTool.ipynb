```python
import subprocess
import time
import re
import math
```

```python
IPs=['239.255.255.250', '192.168.0.255', '104.199.241.202', '104.18.32.47', '224.0.0.1', '224.0.0.252', '224.0.0.22']
hops=30
timeout=1000
```

```python
processes=[]
for ip in IPs:
    processes.append(subprocess.Popen(["tracert","-h",str(hops),"-w",str(timeout),ip],
                                      stdout=subprocess.PIPE,text=True,shell=True))
time.sleep(math.ceil(hops*timeout/1000*3*1.25))
outputs=[]
for process in processes:
    outputs.append(process.stdout)
outputs_trace=[]
for output in outputs:
    temp=[]
    for line in output:
        if line.strip()!='':
            temp.append(line.strip())
    outputs_trace.append(temp)
outputs_trace
```
